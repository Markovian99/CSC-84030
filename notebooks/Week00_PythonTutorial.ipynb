When working on Python projects, especially on your local machine, you'll soon discover that different projects might require different versions of libraries or even Python itself. This is where virtual environments come into play. A virtual environment is an isolated space on your computer that allows you to maintain separate and independent sets of libraries and dependencies for your projects. This isolation prevents conflicts between projects and makes it easier to manage dependencies, ensuring that your projects are reproducible and stable.

#### Using Anaconda or Python's venv

There are several tools available for managing Python virtual environments, but two of the most popular ones are Anaconda and Python's built-in venv module.

**Anaconda**: Anaconda is a tool that aims to simplify package management and deployment. To create a virtual environment in Anaconda, you can use the conda create command followed by the name of the environment and the packages you wish to install. [https://www.anaconda.com/](https://www.anaconda.com/)

**Python venv**: Python's built-in venv module is a great choice and has the advantage of being part of the standard Python library (if you're using Python 3.3 or newer). To create a virtual environment using venv, navigate to your project directory in the terminal and run the following command:

```
python3 -m venv myenv
```

Replace myenv with the name of your virtual environment. Once created, you can activate it using:

On Windows: ```myenv\Scripts\activate.bat```

On Unix or MacOS: ```source myenv/bin/activate```

After activation, any Python or pip commands will only affect the active virtual environment, keeping your global installation clean and your project dependencies isolated.

Setting up a virtual environment might seem like an extra step at the beginning, but it's a best practice that will save you from many headaches as you delve deeper into Python programming. Whether you choose Anaconda or Python's venv, taking the time to manage your environments efficiently will make your coding experience smoother and more enjoyable.

### Functional Programming

Python supports functional programming patterns that can make your code more concise and expressive. Here are some key tools:

````python
# map: apply a function to each element
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x ** 2, numbers))
print("map (square):", squared)

# filter: keep elements that satisfy a condition
evens = list(filter(lambda x: x % 2 == 0, numbers))
print("filter (evens):", evens)
````

````python
# functools.reduce: accumulate values into a single result
from functools import reduce

# sum all numbers
total = reduce(lambda acc, x: acc + x, numbers)
print("reduce (sum):", total)

# product of all numbers
product = reduce(lambda acc, x: acc * x, numbers)
print("reduce (product):", product)
````

````python
# Group by key using a dictionary
# Example: group words by their first letter
words = ["apple", "banana", "avocado", "blueberry", "cherry", "apricot"]

from collections import defaultdict

text = "to be or not to be"
words = text.split()

# Map: emit (word, 1)
mapped = [(w, 1) for w in words]

# Shuffle: group by key
grouped = defaultdict(list)
for k, v in mapped:
    grouped[k].append(v)

# Reduce: sum values per key
reduced = {k: sum(vs) for k, vs in grouped.items()}

print("mapped:", mapped)
print("grouped:", dict(grouped))
print("reduced:", reduced)
````

### Table Arithmetics
- Perform an operation uniformly to all values in a column
- Arithmetics with multiple columns
- Calculate statistics
- Apply a user-defined function to all rows

````python
# Double the extra credits
data = [[60, 70, 80],
        [66, 88, 77],
        [100, 60, 30],
        [85, 87, 83]]
scores = pd.DataFrame(data,
                      index=['Alice', 'Bob', 'Chris', 'David'],
                      columns=['Quiz1', 'Quiz2', 'Final'])
scores['ExtraCredits'] = [1, 2, 3, 4]
print(scores.ExtraCredits)
scores
````

````python
scores.ExtraCredits = scores.ExtraCredits * 2
scores
````

````python
# Calculate grades:
#   Grades = Quiz1 * 25% + Quiz2 * 25% + Final * 50% + ExtraCredit

scores['Grade'] = scores.Quiz1 * 0.25 + scores.Quiz2 * 0.25 + scores.Final * 0.5 + scores.ExtraCredits
scores
````

````python
# Calculate the min, max, mean, median, variance, and std of the final grades

scores['Grade'].max()
scores['Grade'].describe()
````

````python
# Ex: Define a function that determines whether the student passes the class.
def is_passing(grade):
    return (grade >= 60)


# Apply the function to the final grade column
scores['Passed?'] = scores['Grade'].apply(is_passing)
scores
````

````python
def is_passing2(row):
    return (row['Grade'] >= 60)

scores.apply(is_passing2, axis=1)
````

**Box plots** are used for depicting groups of numerical data through their quartiles

- Upper edge: 75% quartile (75% of data are below this value)
- Lower edge: 25% quartile (25% of data are below this value)
- Middle line: median value (50% of data are below this value)
- Upper bar: upper *interquartile range* (values above this bar are considered outliers)
- Lower bar: lower *interquartile range* (values below this bar are considered outliers)
- Dots: outliers
